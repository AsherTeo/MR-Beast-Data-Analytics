# Scaler pre-processing

```python
scaler = StandardScaler()
```

```python
X = df.drop('viewCount_skewed', axis = 1)
y = df[['viewCount_skewed']]
```

```python
X[['likeCount','commentCount','duration(s)','diff_date','Like per day']] = scaler.fit_transform(X[['likeCount','commentCount','duration(s)', 'diff_date','Like per day']])
y['viewCount_skewed'] = scaler.fit_transform(y['viewCount_skewed'].values.reshape(-1, 1))
```

# Cross validation 10-Folds

```python
no_cv = 10

def metrics(cv_mse, cv_mae, cv_r2):
    cv_mse_ave = np.mean(-cv_mse)
    cv_mae_ave = np.mean(-cv_mae)
    cv_r2_ave = np.mean(cv_r2)
    return cv_mse_ave, cv_mae_ave, cv_r2_ave

def cv(model):
    k_folds = KFold(n_splits = no_cv)
    cv_mse = cross_val_score(model, X, y, cv=k_folds, scoring='neg_mean_squared_error')
    cv_mae = cross_val_score(model, X, y, cv=k_folds, scoring='neg_mean_absolute_error')
    cv_r2 = cross_val_score(model, X, y, cv=k_folds, scoring='r2')
    return metrics(cv_mse, cv_mae, cv_r2)
```

```python
random_state = 42


gb = GradientBoostingRegressor()
dt = DecisionTreeRegressor(random_state = random_state)
rf = RandomForestRegressor(random_state = random_state)
et = ExtraTreesRegressor(random_state = random_state)
xgb = XGBRegressor(random_state = random_state)
lgm = LGBMRegressor(random_state = random_state, verbose=-1)

weights = [0.05, 0.025, 0.55, 0.25, 0.075, 0.05]
#ensemble = VotingRegressor(estimators=[('gb', gb), ('dt', dt), ('et', et), ('rf', rf), ('xgb', xgb),('lgm', lgm)])
ensemble = VotingRegressor(estimators=[('gb', gb), ('dt', dt), ('et', et), ('rf', rf), ('xgb', xgb),('lgm', lgm)], weights = weights)
```

# Hyper parameter

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = random_state)
```

## Parameter Tuning

```python
param_grid_dt = {
    'max_depth': [None, 5, 10, 15],  
    'min_samples_leaf': [1, 2, 4],
    'min_samples_split': [2, 5, 10]
    
}

param_grid_gb = {
    'learning_rate' : [0.01, 0.1],
    'n_estimators': [100, 150, 200], 
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

param_grid_rf = {
    'n_estimators': [200, 500], 
    'max_depth': [None, 5, 10, 15],  
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

param_grid_et = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 5, 10, 15],  
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

param_grid_xgb = {'gamma': [0.0468, 0.02,0.01],
                  'learning_rate': [0.05, 0.01, 0.005], 
                  'max_depth':[3, 5, 7], 
                  'n_estimators':[100, 150, 200]
}

param_grid_lgm = {'num_leaves' :[3, 5, 7],
                  'learning_rate':[0.05, 0.02, 0.01], 
                  'n_estimators':[200, 500, 1000],
                  'min_data_in_leaf': [2, 4, 6],
}
```

```python

def best_para(Model, model, param_grid):
    grid_search = GridSearchCV(model, param_grid, scoring='neg_mean_squared_error', cv=no_cv)
    grid_search.fit(X_train, y_train)

    # Get the best hyperparameters
    best_params = grid_search.best_params_

    # Train the model with the best hyperparameters on the entire training set
    best_dt_model = Model(**best_params)
    # Display results
    print("Best Hyperparameters:", best_params)
    return best_params
```

# Decision Tree Hyparameter Tunnning

```python
best_params_rt =  best_para(DecisionTreeRegressor, dt, param_grid_dt)
```

```python
dt_best = DecisionTreeRegressor(max_depth = best_params_rt['max_depth'], 
                                min_samples_leaf = best_params_rt['min_samples_leaf'], 
                                min_samples_split = best_params_rt['min_samples_split'], 
                                random_state = random_state) 

cv_mse_dt_best, cv_mae_dt_best, cv_r2_dt_best = cv(dt_best)
print('MSE:', cv_mse_dt_best)
print('MAE:', cv_mae_dt_best)
print('R2:', cv_r2_dt_best)
```

# Gradient Boosting Hyparameter Tunnning

```python
best_params_gb  =  best_para(GradientBoostingRegressor, gb, param_grid_gb)
```

```python
gb_best = GradientBoostingRegressor(learning_rate = best_params_gb['learning_rate'], 
                                    n_estimators = best_params_gb['n_estimators'], 
                                    min_samples_leaf = best_params_gb['min_samples_leaf'], 
                                    min_samples_split = best_params_gb['min_samples_split'], 
                                    random_state = random_state) 

cv_mse_gb_best, cv_mae_gb_best, cv_r2_gb_best = cv(gb_best)
print('MSE:', cv_mse_gb_best)
print('MAE:', cv_mae_gb_best)
print('R2:', cv_r2_gb_best)
```

# Random Forest Hyparameter Tunnning

```python
best_params_rf = best_para(RandomForestRegressor, rf, param_grid_rf)
```

```python
rf_best = RandomForestRegressor(n_estimators = best_params_gb['n_estimators'], 
                                max_depth = best_params_rf['max_depth'], 
                                min_samples_leaf = best_params_rf['min_samples_leaf'], 
                                min_samples_split = best_params_rf['min_samples_split'], 
                                random_state = random_state) 

cv_mse_rf_best, cv_mae_rf_best, cv_r2_rf_best = cv(rf_best)
print('MSE:', cv_mse_rf_best)
print('MAE:', cv_mae_rf_best)
print('R2:', cv_r2_rf_best)
```

# Extra tree Hyparameter Tunnning

```python
best_params_et  =  best_para(ExtraTreesRegressor, et, param_grid_et)
```

```python
et_best = ExtraTreesRegressor(n_estimators = best_params_et['n_estimators'],
                              max_depth = best_params_et['max_depth'], 
                              min_samples_leaf = best_params_et['min_samples_leaf'], 
                              min_samples_split = best_params_et['min_samples_split'],
                              random_state = random_state) 

cv_mse_et_best, cv_mae_et_best, cv_r2_et_best = cv(et_best)
print('MSE:', cv_mse_et_best)
print('MAE:', cv_mae_et_best)
print('R2:', cv_r2_et_best)
```

# XGB Hyparameter Tunnning

```python
best_params_xgb  =  best_para(XGBRegressor, xgb, param_grid_xgb)
```

```python
xgb_best = XGBRegressor(gamma = best_params_xgb['gamma'], 
                        learning_rate = best_params_xgb['learning_rate'], 
                        max_depth = best_params_xgb['max_depth'], 
                        n_estimators = best_params_xgb['n_estimators'],
                        random_state = random_state) 

cv_mse_xgb_best, cv_mae_xgb_best, cv_r2_xgb_best = cv(xgb_best)
print('MSE:', cv_mse_xgb_best)
print('MAE:', cv_mae_xgb_best)
print('R2:', cv_r2_xgb_best)
```

```python
best_params_lgm  =  best_para(LGBMRegressor, lgm, param_grid_lgm)
```

```python
lgm_best = LGBMRegressor(num_leaves = best_params_lgm['num_leaves'], 
                        learning_rate = best_params_lgm['learning_rate'], 
                        min_data_in_leaf = best_params_lgm['min_data_in_leaf'],
                        n_estimators = best_params_lgm['n_estimators'],
                        random_state = random_state,
                        verbose=-1) 

cv_mse_lgm_best, cv_mae_lgm_best, cv_r2_lgm_best = cv(lgm_best)
print('MSE:', cv_mse_lgm_best)
print('MAE:', cv_mae_lgm_best)
print('R2:', cv_r2_lgm_best)
```

# Testing 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101)
```

```python

y_test_invr = scaler.inverse_transform(y_test)
y_test_org = np.expm1(y_test_invr)
#print(y_test_org)
```

```python
result_df = pd.DataFrame(y_test_org, columns=['testing'])
```

```python
def convert_back(model):
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    pred = pred.reshape(-1, 1)
    pred = scaler.inverse_transform(pred)
    pred = np.expm1(pred)
    return pred
```

```python
dt_model = convert_back(dt_best)
gb_model = convert_back(gb_best)
rf_model = convert_back(rf)
et_model = convert_back(et)
xgb_model = convert_back(xgb_best)
lgm_model = convert_back(lgm)
ensemble_model = convert_back(ensemble_best)
```

```python
result_df['Decision Tree'] = dt_model
result_df['Gradient Bossting'] = gb_model
result_df['Random Forest'] = rf_model
result_df['Extra Tree'] = et_model
result_df['XGB'] = xgb_model
result_df['LGM'] = lgm_model
result_df['Ensemble'] = ensemble_model
```
